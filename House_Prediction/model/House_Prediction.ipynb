```python
import matplotlib.pyplot as plt
%matplotlib inline
def plot_scatter(df,loc):
    bhk2=df[(df['location']==loc) & (df['Rooms']=='2')]
    bhk3=df[(df['location']==loc) & (df['Rooms']=='3')]
    plt.scatter(bhk2.total_sqft,bhk2.price,label="2bhk",color="blue")
    plt.scatter(bhk3.total_sqft,bhk3.price,label="3bhk",color="red",marker='+')
    plt.xlabel("total sqrft ")
    plt.ylabel("Price lakh")
    plt.legend()
    
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor
```

```python
def find_best_model_using_gridsearchcv(X,y):
    algos = {
        'linear_regression' : {
            'model': LinearRegression(),
            'params': {
                'normalize': [True, False]
            }
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1,2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion' : ['mse','friedman_mse'],
                'splitter': ['best','random']
            }
        }
    }
    scores=[]
    for algo_name,M in algos.items():
        cv=ShuffleSplit(n_splits=3,test_size=0.2,random_state=0)
        gds=GridSearchCV(M['model'],M['params'],cv=cv,return_train_score=False)
        gds.fit(X,y)
        scores.append({
            'model':algo_name,
            "best_score":gds.best_score_,
            "best_params":gds.best_params_
        })
    return scores
```

```python
find_best_model_using_gridsearchcv(X,y)
```

```python
def predict_model(location,sqft,bath,balcony,bhk):
    loc_index=np.where(X.columns==location)[0][0]
    x=np.zeros(len(X.columns))
    x[0]=sqft
    x[1]=bath
    x[2]=balcony
    x[3]=bhk
    if loc_index>0:
        x[loc_index]=1
    return model.predict([x])[0]
```

```python
predict_model('2nd Phase Judicial Layout',1000,2,1,5)
```
